```python
from datasets import load_dataset

# Load the dataset
raw_datasets = load_dataset("bentrevett/multi30k")

# Access the splits
train_data = raw_datasets["train"]
val_data = raw_datasets["validation"]
test_data = raw_datasets["test"]

print(f"Train size: {len(train_data)}") # Expected: 29000
```

```python
import spacy
import wandb

# Initialize SpaCy models
spacy_de = spacy.load("de_core_news_sm")
spacy_en = spacy.load("en_core_web_sm")

def tokenize_de(text):
    return [tok.text for tok in spacy_de.tokenizer(text)]

def tokenize_en(text):
    return [tok.text for tok in spacy_en.tokenizer(text)]

# --- W&B Experiment ---
run = wandb.init(project="DA6401_Assignment_3", name="spacy_tokenization_test")

# Create a table to visualize tokenization
test_table = wandb.Table(columns=["Source (DE)", "Target (EN)", "Tokens (DE)", "Tokens (EN)"])

# Test with 5 samples from the training set
for i in range(5):
    de_sentence = train_data[i]['de']
    en_sentence = train_data[i]['en']
    
    test_table.add_data(
        de_sentence, 
        en_sentence, 
        str(tokenize_de(de_sentence)), 
        str(tokenize_en(en_sentence))
    )

# Log the table and finish
run.log({"Tokenization Examples": test_table})
run.finish()
```
